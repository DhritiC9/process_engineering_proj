```python
import networkx as nx
import matplotlib.pyplot as plt

def graphml_to_image(graphml_file_path, output_image_path="flowsheet_graph.png"):
    """
    Reads a GraphML file, creates a networkx graph, and saves it as a PNG image.
    """
    # Step 1: Load the graph
    try:
        G = nx.read_graphml(graphml_file_path)
    except FileNotFoundError:
        print(f"❌ Error: The file '{graphml_file_path}' was not found.")
        return
    except Exception as e:
        print(f"❌ An error occurred while reading the GraphML file: {e}")
        return

    # Step 2: Plot the graph
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, seed=42)  # You can change to kamada_kawai_layout or shell_layout

    labels = {node: data.get('label', node) for node, data in G.nodes(data=True)}

    nx.draw(G, pos, with_labels=False, node_size=3000, node_color="skyblue",
            width=1.5, edge_color="gray", arrows=True, arrowsize=20)
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=12, font_color="black")

    plt.title("GraphML Network Graph")
    plt.axis('off')
    plt.savefig(output_image_path, format="PNG", bbox_inches='tight')
    plt.close()

    print(f"✅ Graph image saved to '{output_image_path}'")
    print(f"📌 Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def display_graphml_inline(graphml_file_path):
    """
    Reads a GraphML file and displays the graph inline (in notebook) using matplotlib.
    """
    try:
        G = nx.read_graphml(graphml_file_path)
    except FileNotFoundError:
        print(f"❌ Error: The file '{graphml_file_path}' was not found.")
        return
    except Exception as e:
        print(f"❌ Error reading GraphML file: {e}")
        return

    # Set layout
    pos = nx.nx_pydot.graphviz_layout(G, prog='dot')
    # Optional: Use 'label' attribute if it exists
    labels = {node: data.get('label', node) for node, data in G.nodes(data=True)}

    # Plot inline
    plt.figure(figsize=(12, 8))
    nx.draw(
        G, pos, with_labels=False,
        node_size=3000, node_color="skyblue",
        edge_color="gray", width=1.5, arrows=True, arrowsize=20
    )
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=12, font_color="black")

    plt.title("GraphML Network Graph")
    plt.axis("off")
    plt.show()

    print(f"📌 Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def display_flowsheet_graph(graphml_file_path):
    try:
        G = nx.read_graphml(graphml_file_path)
    except Exception as e:
        print("❌ Error loading GraphML:", e)
        return

    try:
        # Use Graphviz layout engine via pygraphviz or pydot
        pos = nx.nx_pydot.graphviz_layout(G, prog='dot')    
    except:
        print("⚠️ Falling back to spring layout (Graphviz not installed)")
        pos = nx.spring_layout(G, seed=42)

    # Optional: set node color based on a 'type' attribute, if it exists
    node_colors = []
    for _, data in G.nodes(data=True):
        t = data.get('type', '').lower()
        if 'raw' in t:
            node_colors.append('cornflowerblue')
        elif 'output' in t:
            node_colors.append('lightgreen')
        else:
            node_colors.append('lightgray')

    labels = {node: data.get('label', node) for node, data in G.nodes(data=True)}

    plt.figure(figsize=(14, 8))
    nx.draw(
        G, pos, with_labels=False,
        node_color=node_colors, node_size=3000,
        edge_color='black', arrows=True, arrowsize=20,
        width=1.5
    )
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=10, font_color="black")
    plt.title("Structured Process Flowsheet")
    plt.axis('off')
    plt.show()
display_flowsheet_graph("/Users/dhritichandan/Downloads/SFILES_2/Dhriti_Test/0.graphml")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load graph
G = nx.read_graphml("0.graphml")
G = nx.DiGraph(G)

# Step 1: Apply or rescale positions
pos = nx.spring_layout(G, seed=42)  # You can try shell_layout or others too

# Optional: Scale positions to fit better
for node in pos:
    pos[node] = (pos[node][0]*5, pos[node][1]*5)  # Stretch layout for visibility

# Step 2: Draw the graph
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue')
nx.draw_networkx_edges(G, pos, arrows=True, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

# Optional: Draw edge labels if needed
edge_labels = nx.get_edge_attributes(G, 'label')  # or 'tags'
if edge_labels:
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.axis('off')
plt.title("GraphML Flowsheet Visualization", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Load GraphML file
graphml_path = "0.graphml"  # Update if needed
G = nx.read_graphml(graphml_path)

# Extract node positions from GraphML layout
pos = {}
for node in G.nodes(data=True):
    node_id = node[0]
    data = node[1]

    # Check for yEd-specific position keys
    if 'x' in data and 'y' in data:
        try:
            pos[node_id] = (float(data['x']), -float(data['y']))  # Flip y-axis for Matplotlib
        except:
            continue
    # Check if layout is stored in nested GraphML XML (fallback)
    elif 'graphics' in data:
        graphics = data['graphics']
        pos[node_id] = (float(graphics['x']), -float(graphics['y']))

# Fallback: If no position, use spring layout
if not pos:
    print("No position data found. Using spring_layout.")
    pos = nx.spring_layout(G)

# Plotting
plt.figure(figsize=(10, 10))
nx.draw(
    G,
    pos=pos,
    with_labels=True,
    node_color='skyblue',
    node_size=1000,
    font_size=8,
    font_weight='bold',
    edge_color='gray'
)
plt.title("Graph from .graphml with Layout")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Load GraphML
graphml_path = "0.graphml"
G = nx.read_graphml(graphml_path)

# Extract node positions from attributes (from yEd or similar)
pos = {}
for node_id, data in G.nodes(data=True):
    x = data.get('x')
    y = data.get('y')
    if x is not None and y is not None:
        try:
            pos[node_id] = (float(x), -float(y))  # Flip Y for matplotlib
        except ValueError:
            continue

# If no position info, fallback
if not pos:
    print("No position info found; using spring layout")
    pos = nx.spring_layout(G)

# Set up figure
plt.figure(figsize=(14, 14))

# Draw edges first so they're behind the nodes
nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.4)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=600, node_color='skyblue', alpha=0.9)

# Draw labels with small font and adjusted position
nx.draw_networkx_labels(
    G, pos,
    font_size=7,
    font_weight='bold',
    verticalalignment='center',
    horizontalalignment='center'
)

plt.title("Graph from .graphml with Edges and Layout", fontsize=14)
plt.axis("off")
plt.tight_layout()
plt.show()
```
